```python
pokemons = pd.read_csv('datasets/pokemon.csv', usecols=['attack', 'base_total', 'capture_rate', 'defense', 'height_m', 'hp',
       'name', 'pokedex_number', 'sp_attack', 'sp_defense', 'speed', 'type1',
       'type2', 'weight_kg', 'generation', 'is_legendary'])
pokemons.head()
```

```python
pokemons.tail()
pokemons.columns
pokemons.shape
pokemons.isnull().sum()
pokemons.name
pokemons[['name', 'attack', 'defense']]
pokemons.type1.unique()
pokemons.type1.value_counts()
```

```python
pokemons[pokemons.attack > 100]
pokemons[pokemons.defense < 50]
pokemons[ pokemons['is_legendary'] == 1]
pokemons.loc[ pokemons.generation == 1]
```

```python
pokemons['type2'].isnull().sum()
pokemons.loc[pokemons.type2.isnull()]
pokemons.loc[pokemons["type2"].isnull(), "name"]
pokemons.type2.fillna("None")
pokemons.height_m.fillna(pokemons.height_m.mean())
pokemons.weight_kg.fillna(pokemons.height_m.median())
pokemons.dropna().isna().sum()
```

```python
pokemons.sort_values(by= ["attack"], ascending=True)
pokemons.sort_values(by= ["defense"], ascending=False)
pokemons.sort_values(by = ["speed"], ascending=False)
pokemons.sort_values(by = ["hp"], ascending=False).head()
pokemons.sort_values(by = ['base_total'], ascending=False).iloc[0]
pokemons.sort_values(by=['generation', "attack"])
```

```python
pokemons.groupby('type1').attack.mean()
pokemons.groupby('generation').defense.mean()
pokemons.groupby('generation').size()
# or
pokemons["generation"].value_counts().sort_index()
pokemons.groupby('type1').speed.agg(['max'])
pokemons.groupby('type1').hp.agg(['min'])
pokemons.groupby("type1").weight_kg.mean()
pokemons.groupby('generation').is_legendary.sum()
pokemons.groupby('is_legendary').base_total.mean()
```

```python
pokemons['power'] = pokemons.attack + pokemons.sp_attack
pokemons['bulk'] = pokemons.defense + pokemons.sp_defense
pokemons['is_fast'] = (pokemons.speed > 100).astype(int)
pokemons['is_heavy'] = (pokemons.weight_kg > 100).astype(int)
pokemons
```

```python
pokemons.sort_values(by = ['base_total']).groupby('type1').head(3)
pokemons.sort_values(by=['base_total'], ascending=False).groupby('generation').head(1)
pokemons.loc[pokemons.groupby('type1').attack.idxmax()]
pokemons.groupby('is_legendary')[['attack', 'defense', 'hp', 'speed', 'sp_attack', 'sp_defense']].mean()
pokemons[['attack', 'defense', 'hp', 'speed', 'sp_attack', 'sp_defense']].corr()
pokemons[pokemons.duplicated()]
pd.to_numeric(pokemons.capture_rate, errors= 'coerce')
```

```python
pokemons.groupby('type1').attack.mean().idxmax()
pokemons.groupby('is_legendary').speed.mean().idxmax()
pokemons.groupby('generation').base_total.max().sort_values(ascending=False)
pokemons.groupby('generation').base_total.max().idxmax()
pokemons[["weight_kg", "defense"]].corr()
pokemons.groupby("type1").is_legendary.sum().sort_values(ascending=False)
    #   or
pokemons[pokemons["is_legendary"] == 1]["type1"].value_counts()
pokemons.groupby('type1').base_total.mean().idxmax()
pokemons.groupby("type1")["base_total"].mean().sort_values(ascending=False)
pokemons[pokemons.is_legendary == 0].sort_values(by=['base_total'],ascending=False).head(10)
```

```python
pokemons[[
    "attack","defense","speed",
    "hp","sp_attack","sp_defense"
]].corr()
```

```python
pokemons.sort_values('base_total', ascending=False).groupby("generation").head(5)
num_cols = pokemons.select_dtypes(include=["int64", "float64"]).columns

pokemons[num_cols] = (pokemons[num_cols] - pokemons[num_cols].min()) / \
                    (pokemons[num_cols].max() - pokemons[num_cols].min())
pokemons.pivot_table(values="attack", index="type1", aggfunc="mean")
```
