```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import plotly.express as px
from sklearn.datasets import make_blobs

sns.set_style('whitegrid')
```

```python
hours = [10, 12, 13, 9, 15]
hours.sort()
marks = [40, 43, 45, 38, 50]

plt.plot(hours, marks, marker = 'X')
plt.title("Marks By Hours Studied")
plt.xlabel("Hours")
plt.ylabel("Marks")
plt.show()
```

```python
data = [10, 20, 20, 30, 40, 50, 50, 60]

plt.hist(data, bins=5)
plt.title("Distribution of Values")
plt.xlabel("Value")
plt.ylabel("Frequency")

plt.show()
```

```python
products = ['A', 'B', 'C', 'D']

A_sales = [100, 150, 80, 200]
B_sales = [110, 140, 90, 210]
C_sales = [90, 100, 100, 140]

x = np.arange(len(products))
width = 0.20

plt.bar(x - width, A_sales, width, label = 'A')
plt.bar(x, B_sales, width, label = 'B')
plt.bar(x + width, C_sales, width, label = 'C')

plt.title("Sales by Product", fontsize=14)
plt.xlabel("Products")
plt.ylabel("Sales")
plt.xticks(x, products)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend()

plt.show()
```

```python
products = ['Laptop', 'Mobile', 'Tablet', 'Headphones', 'Smartwatch']

sales_2024 = [50000, 70000, 30000, 20000, 25000]
sales_2025 = [60000, 85000, 40000, 30000, 35000]

x = np.arange(len(products))
width = 0.2

plt.bar(x-width/2, sales_2024, width, label = "2024")
plt.bar(x+width/2, sales_2025, width, label = "2024")

plt.xlabel("Products")
plt.ylabel('Sales')
plt.xticks(x, products)
plt.grid(axis='x', linestyle = '--', alpha = 0.6)
plt.legend()

plt.show()
```

```python
regions = ['North', 'South', 'East', 'West']

profit = [20000, -5000, 15000, 10000]

colors = ['green' if p > 0 else 'red' for p in profit]

plt.bar(regions, profit, color = colors)
plt.title("P/L by Regions")
plt.xlabel('Regions')
plt.ylabel('Profit/Loss')
for i, v in enumerate(profit):
    plt.text(i, v, str(v), ha='center')

plt.axhline(0, color='black')

plt.show()
```

```python
X, y = make_blobs(n_samples= 100, n_features=3, random_state=0)

df = pd.DataFrame(X, columns=['X1', 'X2'])
df['target'] = y
```

```python
fig = px.scatter_3d(data_frame=df, x=df['X1'], y = df['X2'], color = df['target'].astype("str"))

fig.update_traces(marker = dict(size = 12,
                                line = dict(width = 2,
                                            color = 'DarkSlateGrey')),
                  selector=dict(mode = 'markers'))

fig.show()
```
